```python
import pandas as pd

results = pd.DataFrame({
    "Model": [
        "TF-IDF + Naive Bayes",
        "TF-IDF + Logistic Regression",
        "TF-IDF + SVM",
        "BERT + Logistic Regression"],
    "Accuracy": [
        0.93,
        0.9845,
        0.99,
        1.00],
    "Precision": [
        0.93,
        0.98,
        0.99,
        1.00],
    "Recall": [
        0.93,
        0.98,
        0.99,
        1.00],
    "F1 Score": [
        0.93,
        0.98,
        0.99,
        1.00
    ]})
results
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,3))

plt.bar(results["Model"], results["Accuracy"])

plt.xticks(rotation=20)
plt.ylabel("Accuracy")
plt.title("Model Accuracy Comparison")
plt.show()
```

```python
plt.figure(figsize=(5,3))

plt.bar(results["Model"], results["Precision"])

plt.xticks(rotation=20)
plt.ylabel("Precision")
plt.title("Precision Comparison")
plt.show()
```

```python
plt.figure(figsize=(5,3))

plt.bar(results["Model"], results["Recall"])

plt.xticks(rotation=20)
plt.ylabel("Recall")
plt.title("Recall Comparison")
plt.show()
```

```python
plt.figure(figsize=(5,3))

plt.bar(results["Model"] , results["F1 Score"])

plt.xticks(rotation=20)
plt.ylabel("F1 Score")
plt.title("F1 Score Comparison")
plt.show()
```

Note:
The BERT experiment was conducted on a subset of 100 articles due to computational constraints. Therefore, the BERT metrics should not be directly compared with the TF-IDF models evaluated on the full dataset.

The BERT experiment primarily demonstrates the effectiveness of contextual embeddings rather than providing a fully equivalent benchmark.

Observations

1. Naive Bayes achieved the lowest performance among all models.
2. Logistic Regression significantly improved performance over Naive Bayes.
3. SVM achieved the strongest performance among classical machine learning models.
4. BERT embeddings combined with Logistic Regression achieved perfect accuracy on the selected subset.
5. The BERT experiment used a much smaller dataset and therefore results should be interpreted cautiously.
6. Contextual embeddings generated by BERT capture semantic information more effectively than TF-IDF representations.

```python
cost_df = pd.DataFrame({
    "Model": [
        "Naive Bayes",
        "Logistic Regression",
        "SVM",
        "BERT + Logistic Regression"],

    "Training Cost": [
        "Very Low",
        "Low",
        "Medium",
        "High"],

    "Inference Cost": [
        "Very Low",
        "Low",
        "Medium",
        "High"]
})

cost_df
```

Accuracy vs Computational Cost

Naive Bayes offers very fast training and inference but lower performance.

Logistic Regression provides a strong balance between accuracy and efficiency.

SVM achieved the best classical NLP performance while remaining computationally manageable.

BERT generated the strongest semantic representations but required substantially greater computational resources.

```python
best_model = results.loc[results["Accuracy"].idxmax()]

print(best_model)
print("\nBest Performing Model:")
print(best_model["Model"])
```

Conclusion

Classical machine learning methods such as Logistic Regression and SVM performed strongly on TF-IDF representations.

BERT embeddings further improved semantic understanding of news articles.

Among classical methods, SVM achieved the highest accuracy.

BERT-based representations showed the strongest overall performance on the evaluated subset.

Future work will focus on explainability, uncertainty estimation, and advanced multimodal techniques.
